```python
import numpy as np
import math
import sys
import matplotlib.pyplot as plt
import copy
```

```python
X1 = [1, 1.5, 5, 3, 4, 3]
n = len(X1)
size = len(X1)
cluster = []
clusterm = []
cluster_idx = []
dist_mat = []
count = 1

a = -1
b = -1
```

```python
for i in range (n):
  m = []
  cluster_idx.append([X1[i]])
  for j in range (n):
    val1 = math.pow(X1[i],2)
    val2 = math.pow(X1[j],2) 
    if (i == j) :
      d = math.inf
      
    else : 
      if (val1>val2):
        d = math.sqrt(val1-val2)
      else:
        d = math.sqrt(val2-val1)
    m.append(d)
  dist_mat.append(m)

print(dist_mat)
      
```

```python
cluster.append(dist_mat)
clusterm.append(cluster_idx)

while count < size:
  dist_mat = cluster[count-1]
  min_dist = sys.float_info.max 
  for i in range(0,n-1):
    for j in range(i+1, n):
      d = dist_mat[i][j]
      if min_dist >= d:
        min_dist = d
        a = i
        b = j
      
    
  print(min_dist)
  cluster_idx = copy.deepcopy(clusterm[count-1])
  for i in range(0,len(cluster_idx[b])):
    cluster_idx[a].append(cluster_idx[b][i])
  cluster_idx.pop(b)
 

  for i in range(0, n):
    dist_mat[i][a] = dist_mat[i][a] if dist_mat[i][a]<dist_mat[i][b] else dist_mat[i][b];

  dist_mat = np.delete(np.delete(dist_mat, b, 0),b,1)
  
  n = n-1
  clusterm.append(cluster_idx)
  cluster.append(dist_mat)

  count+=1

print(cluster)
```
