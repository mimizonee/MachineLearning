```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([2, 6, 5, 7])
Y = np.array([3, 10, 4, 13])
w = [0, 0] #starting with random value
lr = 0.0001
epoch = 1000

mean_X = np.mean(X)
mean_Y = np.mean(Y)
n = len(X)

cost_data = []
epoch_data = []
```

Visualising the Data 

```python
plt.scatter(X,Y, c = "red",alpha=0.5,marker = 'o')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Linear Model

```python
def h(X, w):
  coeff_w0 = [1, 1, 1, 1]
  coeff_w1 = X
  a = np.array([coeff_w0, coeff_w1])
  pred = np.matmul(w, a)
  return(pred)
```

Cost Function

```python
def cost(w, X, Y):
  J = (0.5/n)*np.sum(np.square(h(X, w)-Y))
  return J
```

Gradient Descent

```python
def grad(w, X, Y):
  dw0 = (1/n)*np.sum(h(X, w)-Y)
  dw1 = (1/n)*np.sum((h(X, w)-Y)*X)
  g = np.array([dw0, dw1])
  return g
```

```python
def update(w_new, w_prev, lr, e):
  epoch = e
  print('weight = ', w_prev )
  print('J = ', cost(w_prev, X, Y))
  for i in range(0,epoch):
    w_new[0] = w_prev[0] - lr*grad(w_prev,X,Y)[0]
    w_new[1] = w_prev[1] - lr*grad(w_prev,X,Y)[1]

    print('weights = ', w_new )
    print('J = ', cost(w_new,X,Y))

    cost_data.append(cost(w_new,X,Y))
    epoch_data.append(i)

    if ((((w_new[0]-w_prev[0]) < 0.001) and ((w_new[0]-w_prev[0]) > 0.001)) and (((w_new[1]-w_prev[1]) < 0.001) and ((w_new[1]-w_prev[1]) > 0.001))):
   #if ((cost(w_prev, X, Y) - cost(w_new, X, Y)) < 0.001):
      return w_new
    w_prev = w_new
    #epoch = epoch-1

  return w_new



```

Training the model

```python
w = [0, 0]
w = update(w, w, lr, epoch)
print(w)
```

```python
def graph(formula, x_range):  
    x = np.array(x_range)  
    y = formula(x)  
    plt.plot(x, y)  
    
def my_formula(x):
    return w[0]+w[1]*x

plt.scatter(X,Y, c = "red",alpha=.5, marker = 'o')
graph(my_formula, range(-10,10))
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.scatter(epoch_data,cost_data, c = "red",alpha=.5, marker = 'o')
plt.xlabel('Epoch')
plt.ylabel('Cost')
plt.show()
```
